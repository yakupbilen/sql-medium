```python
%%sql
-- Maçlarda atılan gollerin tarihe göre kümülatif toplamı
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    SUM(away_team_goal+home_team_goal) OVER(ORDER BY date) AS cum_sum
FROM match
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin tarihe göre hareketli ortalaması
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    ROUND(AVG(away_team_goal+home_team_goal) OVER(ORDER BY date),2) AS cum_avg
FROM match
LIMIT 20;
```

```python
%%sql
-- Tarihe göre, o zamana dek bir maçta atılan en fazla gol
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    MAX(away_team_goal+home_team_goal) OVER(ORDER BY date) AS mov_max
FROM match
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin aylara(partition by month) göre kümülatif bir şekilde toplanması
-- Aşağıdaki sorgunun önceki sorgu ile arasındaki farkı 15.satıra bakarak anlayabilirsiniz.
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    SUM(away_team_goal+home_team_goal) OVER(PARTITION BY EXTRACT(MONTH from date) ORDER BY date) AS cum_sum
FROM match
ORDER BY date,id
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin aylara göre hareketli ortalaması
-- Aşağıdaki sorgunun önceki sorgu ile arasındaki farkı 15.satıra bakarak anlayabilirsiniz.
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    ROUND(AVG(away_team_goal+home_team_goal) OVER(PARTITION BY EXTRACT(MONTH from date) ORDER BY date),2) AS cum_avg
FROM match
ORDER BY date,id
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin belirtilen çerçeveye göre toplanması.
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    SUM(away_team_goal+home_team_goal) OVER(ORDER BY date 
                ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS frame_sum
FROM match
ORDER BY date
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin belirtilen çerçeveye göre toplanması.
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    SUM(away_team_goal+home_team_goal) OVER(ORDER BY date 
                ROWS BETWEEN CURRENT ROW AND UNBOUNDED FOLLOWING) AS frame_sum
FROM match
ORDER BY date DESC,frame_sum ASC
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin belirtilen çerçeveye göre toplanması.
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    SUM(away_team_goal+home_team_goal) OVER(ORDER BY date 
                ROWS BETWEEN 2 PRECEDING AND CURRENT ROW) AS frame_sum
FROM match
ORDER BY date
LIMIT 20;
```

```python
%%sql
-- Maçlarda atılan gollerin belirtilen çerçeveye göre toplanması.
SELECT
    date,
    away_team_goal+home_team_goal AS total_goal,
    SUM(away_team_goal+home_team_goal) OVER(ORDER BY date 
                ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING) AS frame_sum
FROM match
ORDER BY date
LIMIT 20;
```
