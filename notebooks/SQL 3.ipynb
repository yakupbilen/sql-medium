```python
%%sql
-- 2010 yılında, hayat beklentisinin en düşük olduğu 5 ülke (ülke ismi,hayat beklentisi)
SELECT
    c.name AS "Country Name",
    p.life_expectancy AS "Life Expectancy"
    FROM countries AS c
        INNER JOIN populations AS p
        ON c.code=p.country_code
WHERE p.year = 2010
ORDER BY "Life Expectancy"
LIMIT 5;
```

```python
%%sql
-- Güney Avrupa'daki ülkelerin kullandığı para birimleri
SELECT name AS country, basic_unit
FROM countries
  INNER JOIN currencies
    USING (code)
WHERE region = 'Southern Europe'
ORDER BY region;
```

```python
%%sql
-- 2010 yılındaki kıtaların nüfusları
SELECT
    c.continent AS "Continent",
    SUM(p.size) AS "Total Population"
    FROM countries_plus AS c
    LEFT JOIN populations AS p
    ON c.code = p.country_code
WHERE p.year=2010
GROUP BY "Continent"
ORDER BY "Total Population" DESC;
```

```python
%%sql
-- 2010 yılındaki kıtaların nüfusları
SELECT
    c.continent AS "Continent",
    SUM(p.size) AS "Total Population"
    FROM populations AS p
    RIGHT JOIN countries_plus AS c
    ON c.code = p.country_code
WHERE p.year=2010
GROUP BY "Continent"
ORDER BY "Total Population" DESC;
```

```python
%%sql
-- Kuzey Amerikada veya hangi kıtada olduğu bilinmeyen ülkelerin,isimleri,ülke kodları,bulunduğu kıta ve para birimleri
SELECT name AS country, code, region, basic_unit
FROM countries
  FULL JOIN currencies
    USING (code)
WHERE region = 'North America' OR region IS null 
ORDER BY region;
```

```python
%%sql
-- Suriye'nin 2010 ve 2015 yıllarındaki enflasyonu
SELECT 
    c.name,
    e.year,
    e.inflation_rate
    FROM economies AS e
    FULL JOIN countries AS c
    USING(code)
WHERE c.name = 'Syria';
```

```python
%%sql

SELECT 
    c.name AS city,
    l.name AS language
    FROM cities AS c        
    CROSS JOIN languages AS l
WHERE c.name = 'London'
LIMIT 10;
```
